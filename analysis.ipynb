```python
plt.figure(figsize=(6,4))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5, color="Green")
plt.xlabel('Living Area (sqft)')
plt.ylabel('House Price')
plt.title('House Price vs Living Area')
plt.show()
```

- The scatter plot shows a moderate positive relationship between living area and house price. 
- While prices tend to increase with square footage, there is considerable variability - homes of similar size can have vastly different prices

```python
fig, axes = plt.subplots(1,2, figsize=(10,4))

x1 = df['condition']
y1 = df['price']

x2 = df['sqft_living']
y2 = df['price']

axes[0].scatter(x1,y1)
axes[0].set_title("Plot 1")
axes[0].set_xlabel("Condition")
axes[0].set_ylabel("Price")


axes[1].scatter(x2,y2)
axes[1].set_title("Plot 2")
axes[1].set_xlabel("House Size Sqft")
axes[1].set_ylabel("Price")


plt.tight_layout()
plt.show()
```

- House condition shows a weak relationship with price, as prices vary widely across all condition levels. 
- In contrast, house size demonstrates a strong positive relationship with price, indicating that size plays a much more significant role in determining house value than condition.

```python
plt.scatter(df['bedrooms'], df['price'], alpha=0.5)

plt.title('Bedrooms Vs Prices')
plt.xlabel('Bedrooms')
plt.ylabel('Prices')

plt.show()
```

- `plt.scatter` plot is best for diminishing returns.
- Beyond 4â€“5 bedrooms, price variance increases, indicating that other factors (location, sqft, condition) dominate pricing rather than bedroom count alone

```python
fig, axes = plt.subplots(1,2, figsize=(10,4))

x1 = df['bathrooms']
y1 = df['price']

x2 = df['bedrooms']
y2 = df['price']

axes[0].scatter(x1,y1, color='Red')
axes[0].set_title("Plot 1")
axes[0].set_xlabel("Bathrooms")
axes[0].set_ylabel("Price")


axes[1].scatter(x2,y2, color='Green')
axes[1].set_title("Plot 2")
axes[1].set_xlabel("Bedrooms")
axes[1].set_ylabel("Price")


plt.tight_layout()
plt.show()
```

- Bathrooms show a stronger and more consistent relationship with price than bedrooms.
- Bedrooms exhibit diminishing returns, while bathrooms continue to add value.

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].scatter(df['condition'], df['price'], alpha=0.5, color='green')
axes[0].set_title('Condition Vs Price')
axes[0].set_xlabel('Condition')
axes[0].set_ylabel('Price')

axes[1].scatter(df['sqft_living'], df['price'], alpha=0.5, color='skyblue', edgecolor='Black')
axes[1].set_title('Size Vs Price')
axes[1].set_xlabel('Size Sqft')
axes[1].set_ylabel('Price')


plt.tight_layout()
plt.show()

#Checking Correlation

cond_corr = df['condition'].corr(df['price'])
size_corr = df['sqft_living'].corr(df['price'])

print("Correlation (Condition vs Price):", cond_corr)
print("Correlation (Size vs Price):", size_corr)
```

- House size has a significantly stronger impact on price than house condition, with a correlation of approximately 0.43 for living area, compared to nearly 0.03 for condition. 
- This suggests that larger houses are generally priced higher, while condition does not strongly affect pricing.
